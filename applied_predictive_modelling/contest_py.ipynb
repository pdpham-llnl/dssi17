```python
from IPython.display import display

import sklearn.datasets as ds
import matplotlib as plt
%matplotlib inline
```

```python
import pandas as pd
from pandas import read_csv

MAX_ROW = 12
pd.options.display.max_rows=MAX_ROW
```

```python
Xtrain = read_csv('cars_Xtrain.csv')
Ytrain = read_csv('cars_Ytrain.csv')

Xtest = read_csv('cars_Xtest.csv')
Ytest = read_csv('cars_Ytest.csv')
```

```python
def view_all(df): 
    pd.options.display.max_rows=df.shape[0]
    display(df)
    pd.options.display.max_rows=MAX_ROW
view_all(Xtrain)
```

```python
for c in ['displacement','horsepower','weight','acceleration']:
    Xtrain.hist(column=c,bins=100)
```

```python
category_columns = ['cylinders','year','origin','make','model']
for cc in category_columns:
    view_all(Xtrain.groupby(cc).size())
```

Toy program fit ARIMA model with U.S. GNP data

```r
library(astsa)
```

```r
plot(gnp)
acf2(gnp, 50)
gnpgr = diff(log(gnp))      # growth rate
plot(gnpgr)
acf2(gnpgr, 24)  
sarima(gnpgr, 1, 0, 0)      # AR(1)
sarima(gnpgr, 0, 0, 2)      # MA(2) 
ARMAtoMA(ar=.35, ma=0, 10)  # prints psi-weights
```

```python
import pandas as pd
import numpy as np
from pandas_datareader.data import DataReader
from datetime import datetime

import matplotlib.pylab as plt
plt.style.use('ggplot')

%matplotlib inline

a = DataReader('JPM',  'google', datetime(2006,6,1), datetime(2016,6,1),)
a_returns = pd.DataFrame(np.diff(np.log(a['Close'].values)))
a_returns.index = a.index.values[1:a.index.values.shape[0]]
a_returns.columns = ["JPM Returns"]

a_returns.head()
```

```python
plt.figure(figsize=(15, 5))
plt.ylabel("Returns")
plt.plot(a_returns)
plt.show()
```

```python
from statsmodels.tsa.stattools import acf,pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(a_returns.values.T[0],lags=48,)
plot_pacf(a_returns.values.T[0],lags=48,)
```
